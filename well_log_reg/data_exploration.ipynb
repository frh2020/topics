# Quick data exploration

Data exploration with AutoViz

```
pip install autoviz
```

```
import pandas as pd
import numpy as np

from autoviz.AutoViz_Class import AutoViz_Class
```

```
data = pd.read_csv('train_wells_20210131_ps1.csv')
data.shape
```

```
AV = AutoViz_Class()
```

```
filename = ""
depVar = "DTSM"
dfte = data
max_rows = max(300000, data.shape[0])
max_cols = max(30, data.shape[1])

AV.AutoViz(
    filename,
    sep=",",
    depVar=depVar,
    dfte=dfte,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=max_rows,
    max_cols_analyzed=max_cols
)
```
